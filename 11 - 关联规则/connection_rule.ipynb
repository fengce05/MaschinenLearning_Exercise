```python
import pandas as pd
import mlxtend
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
data ={'ID':[1,2,3,4,5,6],
       'Onion':[1,0,0,1,1,1],
       'Potato':[1,1,0,1,1,1],
       'Burger':[1,1,0,0,1,1],
       'Milk':[0,1,1,1,0,1],
       'Beer':[0,0,1,0,1,0]
       }
```

```python
df=pd.DataFrame(data)
```

```python
df =df[['ID','Onion','Potato','Burger','Milk','Beer']]
```

```python
df
```

```python
frequent_itemsets = apriori(df[['Onion','Potato','Burger','Milk','Beer']],min_support=0.5,use_colnames=True)
```

```python
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets,metric='lift',min_threshold=1,num_itemsets=10)
```

```python
rules
```

```python
rules[(rules['lift']>1.125)&(rules['confidence']>0.8)]
```

```python
retails_shopping_basket = {'ID':[1,2,3,4,5,6],
                           'Basket':[['Beer','Diaper','Pretzels','Chips','Aspirin'],
                                     ['Diaper','Beer','Chips','Lotion','Juice','Babyfood','Milk'],
                                     ['Soda','Chips','Milk'],
                                     ['Soup','Beer','Diaper','Milk','IceCream'],
                                     ['Soda','Coffee','Milk','Bread'],
                                     ['Beer','Chips']
                                     ]
                           }
```

```python
retail = pd.DataFrame(retails_shopping_basket)
```

```python
retail_id = retail.drop('Basket', axis=1)
retail_id
```

```python
retail_Basket = retail_Basket.str.get_dummies(',')
retail_Basket
```
